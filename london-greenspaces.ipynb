# London Green Spaces Accessibility Analysis

This file focuses on analysing the accessibility of green spaces in London by examining the spatial relationship between green spaces and the road network. It aims to assess how well-connected London's green spaces are to the road infrastructure, providing insights into urban planning and sustainability.

I will be using the data I have cleaned inside the `london-road-merge.ipynb` file, including:
- London boundary
- London green space
- London roads


```python
# Import libraries to read and load geospatial files
import geopandas as gpd
```

```python
# Loading datasets
london_polygon = gpd.read_file("Data/clean-data/london.gpkg")
london_green_space = gpd.read_file("Data/clean-data/london_green_space.gpkg")
london_roads = gpd.read_file("Data/clean-data/london_roads.gpkg")
```

## Original map plots

```python
# Plot new London green space data
green_space_plot = london_green_space.plot(color='green')
green_space_plot.set(xlabel="Easting", ylabel="Northing", title="London Green Spaces")
```

```python
# Plotting London roads to see road types
london_road_plot = london_roads.plot(column='form_of_way', 
                                     legend=True,
                                     legend_kwds={'loc':'lower left',
                                                  'fontsize':8})
london_road_plot.set(xlabel="Easting", ylabel="Northing", title="London Road Types")
```
